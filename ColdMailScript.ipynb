```python
import csv

def generate_email(name, email_format, company, domain):
    # Split the name into first name and last name
    first_name, last_name = name.split()

    # Replace placeholders in the email format with actual values
    email = email_format.replace('[First Name]', first_name)
    email = email.replace('[Company Name]', company)
    email = email.replace('[Company Name]', company)

    return email

names = ['John Doe', 'Jane Smith', 'Michael Johnson']  # Example list of names
company = 'microsoft'  # Example company name
domain = 'example.com'  # Example domain
email_format = "Subject: Could I be your next intern?\n\nDear [First Name],\n\nI am Krishnna Sarrdah, an MS student at Northeastern University, with a keen interest in software engineering. I am reaching out to inquire about potential internship opportunities at [Company Name] (I would also love to be considered for full-time entry-level positions starting summer 24).\n\nMy background includes hands-on programming skills and theoretical understanding from my ongoing Master's in Information Systems and a Bachelor's degree in Electronics and Telecommunication. My prior experience as a Full Stack Software Development Intern at Perkeasy has helped me gain a practical understanding of software development and team collaboration.\n\nI am excited by the innovative work [Company Name] is doing and I genuinely believe my hard work and grit can make a difference there.\n\nI have attached my resume for your convenience. Please feel free to contact me via email at sarrdah.k@northeastern.edu\n\nThank you for considering my application.\nBest Regards,\nKrishnna Sarrdah"

rows = []
for name in names:
    email = generate_email(name, email_format, company, domain)
    rows.append([company, name, email])
    print(f"{company} - {name}: {email}")
    print("\n")

filename = f"{company.lower().replace(' ', '_')}_employee_data.csv"
with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Company', 'Name', 'Email'])
    writer.writerows(rows)

print(f"Employee data for {company} has been stored in '{filename}'.")
```
